```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
a=pd.read_csv('Titanic-Dataset.csv')
e=a[['Age','Sex']].dropna(subset=['Age'])
plt.figure(figsize=(10, 6))
sns.histplot(data=e, x='Age', hue='Sex',multiple='stack' , bins=30, palette='Set2')

# Add labels and title
plt.title('Age Distribution of Titanic Passengers by Sex')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
